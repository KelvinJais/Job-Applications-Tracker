```python
def read_text_from_file(file_path):
    try:
        # Open the file in read mode
        with open(file_path, 'r') as file:
            # Read the entire content of the file
            text = file.read()
        return text
    except FileNotFoundError:
        return "File not found. Please check the file path."
    except Exception as e:
        return f"An error occurred: {e}"

# Example usage
file_path = 'applied.txt'  # Replace with your file path
applied = read_text_from_file(file_path)
reject=read_text_from_file("reject.txt")

all_mail=reject+applied
```

```python
import spacy
from collections import Counter
nlp = spacy.load('en_core_web_sm')
doc = nlp(all_mail)
# Extract lemmatized words
lemmatized_words = [token.lemma_ for token in doc if token.is_alpha and not token.is_stop]
# Count the frequency of each lemmatized word
word_freq = Counter(lemmatized_words)
```
